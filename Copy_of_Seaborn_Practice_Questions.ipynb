## Question 3: Distribution of Total Fare Amounts

Problem Statement:
We want to understand how total fare amounts are distributed among taxi trips.

Approach:
1. Use the 'fare' column from the 'taxis' dataset.
2. Plot a histogram or KDE plot using `sns.histplot()`.
3. Add labels and title to interpret the data better.

```python
# Write your code here
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

taxis = sns.load_dataset('taxis')

plt.figure(figsize=(8, 5))
sns.histplot(data=taxis, x='fare', bins=30, kde=True, color='skyblue')

plt.title('Distribution of Total Fare Amounts', fontsize=14)
plt.xlabel('Total Fare Amount ($)', fontsize=12)
plt.ylabel('Number of Trips', fontsize=12)

plt.show()
```

## Question 4: Relationship Between Distance and Fare

## Question 5: Average Fare by Day of the Week

Problem Statement:
Do taxi fares vary across different days of the week?

Approach:
1. Use the 'pickup_day' or derive day information from the datetime column if available.
2. Plot the average fare per day using `sns.barplot()`.

```python
# Write your code here
taxis = sns.load_dataset('taxis')

taxis['pickup_day'] = taxis['pickup'].dt.day_name()

plt.figure(figsize=(8, 5))
sns.barplot(data=taxis, x='pickup_day', y='fare', estimator='mean', palette='cool')

plt.title('Average Taxi Fare by Day of the Week', fontsize=14)
plt.xlabel('Day of the Week', fontsize=12)
plt.ylabel('Average Fare ($)', fontsize=12)

plt.show()
```

## Question 8: Trip Duration vs. Fare Amount

Problem Statement:
Is there any relationship between how long a trip takes and how much it costs?

Approach:
1. Use columns like 'pickup' and 'dropoff' to calculate trip duration (if timestamps are available).
2. Plot a scatter plot between duration and fare using `sns.scatterplot()`.

```python
# Write your code here
taxis['trip_duration'] = (taxis['dropoff'] - taxis['pickup']).dt.total_seconds() / 60

plt.figure(figsize=(8, 5))
sns.scatterplot(data=taxis, x='trip_duration', y='fare', color='teal')

plt.title('Relationship Between Trip Duration and Fare', fontsize=14)
plt.xlabel('Trip Duration (minutes)', fontsize=12)
plt.ylabel('Fare Amount ($)', fontsize=12)

plt.show()
```

## Question 10: Average Tip by Distance Category

Problem Statement:
We want to know if longer trips generally result in higher tips.

Approach:
1. Create distance categories (e.g., short, medium, long) using the 'distance' column.
2. Use `sns.barplot()` to plot average tips for each distance category.

```python
# Write your code here

bins = [0, 2, 5, taxis['distance'].max()]
labels = ['Short', 'Medium', 'Long']
taxis['distance_category'] = pd.cut(taxis['distance'], bins=bins, labels=labels, include_lowest=True)

plt.figure(figsize=(8, 5))
sns.barplot(data=taxis, x='distance_category', y='tip', estimator='mean', palette='coolwarm')

plt.title('Average Tip by Distance Category', fontsize=14)
plt.xlabel('Distance Category', fontsize=12)
plt.ylabel('Average Tip Amount ($)', fontsize=12)

plt.show()
```
